```python
# Create quantum circuit with 4 qubits and 2 classical bits
qc = QuantumCircuit(4, 2)

# qubit 0 and 1 are negated (set to 1 instead of 0)
qc.x(0)
qc.x(1)

# CNOT controlled by qubit 1 and targeting qubit 2
qc.cx(1,2)

# CNOT controlled by qubit 0 and targeting qubit 1
qc.cx(0,2)

# qubit 3 is the sum of qubit 0 and qubit 1
qc.ccx(0,1,3)

# Measure
qc.measure(2,0)
qc.measure(3,1)

# Display a drawing of the circuit
display(qc.draw())

# Run the experiment
job = sim.run(qc)

# Get the results
result = job.result()

# Interpret the results as a "counts" dictionary
result.get_counts()
```

```python
from qiskit.visualization import plot_histogram

# Plot the histogram of the results
plot_histogram(result.get_counts())
```

```python
# Create quantum circuit with 2 qubits and 2 classical bits
qc2 = QuantumCircuit(2, 2)

# Create a Hadamard gate
qc2.h(0)

# Measure
qc2.measure(0,0)
qc2.measure(1,1)

# Display a drawing of the circuit
display(qc2.draw())

# Run the experiment
job2 = sim.run(qc2)

# Get the results
result2 = job2.result()

# Interpret the results as a "counts" dictionary
result2.get_counts()
```

```python
backend = Aer.get_backend('statevector_simulator')

# qubit 1 is negated (set to 1 instead of 0)
qc2.x(1)

# Run the experiment
job3= backend.run(qc2)

# Get the results
result3=job3.result()
outputstate = result3.get_statevector(qc2,decimals=3)
print(outputstate)
```
